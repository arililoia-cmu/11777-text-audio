```python
from bert_score import score
import pandas as pd
import csv
```

```python
csv_path = '/content/lpmc/test_set_captioning.csv'
df = pd.read_csv(csv_path)
track_id = df['track_id'].tolist()[1:]
generated_caption = df['generated_caption'].tolist()[1:]
tag = df['tag'].tolist()[1:]
caption_writing = df['caption_writing'].tolist()[1:]
caption_summary = df['caption_summary'].tolist()[1:]
caption_paraphrase = df['caption_paraphrase'].tolist()[1:]
caption_attribute_prediction = df['caption_attribute_prediction'].tolist()[1:]
```

```python
#generated caption P, R, F1
tag_P, tag_R, tag_F1 = score(generated_caption, tag, lang='en', verbose=True)
writing_P, writing_R, writing_F1 = score(generated_caption, caption_writing, lang='en', verbose=True)
summary_P, summary_R, summary_F1 = score(generated_caption, caption_summary, lang='en', verbose=True)
paraphrase_P, paraphrase_R, paraphrase_F1 = score(generated_caption, caption_paraphrase, lang='en', verbose=True)
attrib_pred_P, attrib_pred_R, attrib_pred_F1 = score(generated_caption, caption_attribute_prediction, lang='en', verbose=True)
```

```python
tags_set = set()
tags_list = []
# for i in range(len(tag)):
for comma_separated_string in tag:
    # Split the comma-separated string into a list of strings
    split_strings = comma_separated_string.split(',')
    # Add each string to the set
    tags_set.update(split_strings)
    tags_list.append(set(split_strings))

# unique_tags_list - list of all unique tags in dataset
unique_tags_list = list(tags_set)
# tags_list: list of sets of tags for audio clips
print(unique_tags_list)
```

```python
def get_indices_with_tag(tag_):
  indices = []
  for i in range(len(tags_list)):
    if tag_ in tags_list[i]:
      indices.append(i)
  return indices
```
