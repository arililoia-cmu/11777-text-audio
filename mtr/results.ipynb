```python
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import json

with open("ecals_tags.json") as infile:
    tags = json.load(infile)
    infile.close()
```

```python
plt.title("Histogram of ECALS Tag Frequencies")
plt.xlabel("Tag Frequency")
plt.ylabel("Count")
plt.hist(tags.values(), bins=30)
plt.show()
```

```python
with open("50_results.json") as infile:
    results_50 = json.load(infile)
    infile.close()

with open("959_results.json") as infile:
    results_959 = json.load(infile)
    infile.close()
```

```python
roc_auc_values_50 = {}
pr_auc_values_50 = {}

for tag in results_50["tag_wise"]:
    roc_auc_values_50[tag] = results_50["tag_wise"][tag]["roc_auc"]
    pr_auc_values_50[tag] = results_50["tag_wise"][tag]["pr_auc"]
```

```python
np.mean(list(roc_auc_values_50.values())), np.std(list(roc_auc_values_50.values()))
```

```python
np.mean(list(pr_auc_values_50.values())), np.std(list(pr_auc_values_50.values()))
```

```python
plt.title("Histogram of AUC values for 50 ECALS Tags")
plt.xlabel("AUC Value")
plt.ylabel("Count")
plt.hist(roc_auc_values_50.values(), label="ROC", color="cyan")
plt.hist(pr_auc_values_50.values(), label="PR", color="skyblue")
plt.legend()
plt.show()
```

```python
roc_auc_values_959 = {}
pr_auc_values_959 = {}

for tag in results_959["tag_wise"]:
    roc_auc_values_959[tag] = results_959["tag_wise"][tag]["roc_auc"]
    pr_auc_values_959[tag] = results_959["tag_wise"][tag]["pr_auc"]
```
